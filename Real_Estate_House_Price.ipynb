```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/housing_data.csv')
```

```python
df.info()
```

```python
# Identify numerical columns: columns with more than 5 unique values are considered numerical
numerical_columns = [col for col in df.columns if df[col].nunique() > 5]

# Identify categorical columns: columns that are not numerical and not 'GradeClass'
categorical_columns = df.columns.difference(numerical_columns).difference(['SalePrice']).to_list()
```

```python
for column in categorical_columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=column)
    plt.title(f'Countplot of {column}')

    # Directly set custom labels
    #labels = categorical_columns[column]
    #ticks = range(len(labels))
    #plt.xticks(ticks=ticks, labels=labels)

    plt.tight_layout()
    plt.show()
```
