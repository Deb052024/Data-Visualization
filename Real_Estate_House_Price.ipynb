**Make Feature enginerring Function To make One column from two columns**

```python
def column_addition(x,y):
  if x!=y:
    return (x+' &' +y)
  else:
      return x
```

```python
df.columns
```

```python
df.duplicated().sum()
```

**"Feature Engineering" : Addition of Two columns to One**

```python
df['Condition1']=df.apply(lambda a:column_addition(a['Condition1'],a['Condition2']),axis=1)
```
